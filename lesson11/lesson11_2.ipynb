```python
import numpy as np
from numpy import ndarray

students:ndarray = np.array([[12,55,33,68,50],
                           [68,88,77,66,55]])
#利用索引，slicing
students[0] 
#取得第一位學生資料。第一項數字表示第一維資料(學生)
students[1] 
#取得第二位學生資料
students[1,:3] 
#取得第二位學生，前三科分數。第二項數字表示第二維資料（學生分數），冒號表示起始值：結束值：間隔
students[0,-3:] 
#取得第一位學生資料，後三科分數
```

# 繪製圖表

```python
subjects:list[str] = ['chinese', 'english', 'math', 'history', 'discover']
subjects
```

```python
import matplotlib.pyplot as plt

figure = plt.figure() 
#畫布 這行會建立一個空白的繪圖「畫布」（Figure）。
#Figure 是 Matplotlib 中最上層的容器，所有的圖表元素（如座標軸、標題、圖例等）都會放在這個畫布裡。
#你可以把它想像成一張空白的畫紙。
axes = figure.add_subplot()
#這行是在剛剛建立的畫布上新增一個「子圖」（subplot），也就是一個繪圖區域（Axes）。
#如果沒有指定參數，預設會建立一個 1x1 的子圖，也就是整個畫布只有這一個繪圖區域。
#這個子圖物件負責繪製實際的圖形（如折線圖、長條圖等），並管理座標軸、標籤等細節。

axes.plot(subjects, students[0],label="stu1") #生成第一條折線圖
axes.plot(subjects,students[1],label="stu2") #生成第二條折線圖
axes.legend() #顯示圖例
axes.set_title("stu1 vs stu2",fontsize=24,color="r") #設定標題
axes.set_xlabel("subject",fontsize=16) # 設定X軸資料名稱
axes.set_ylabel('scores') #設定Y軸資料名稱
plt.savefig("test.jpeg") #儲存圖檔，生成檔案
plt.show() #顯示圖表
plt.close() #關閉
```

# 改成繪製長條圖

```python
import numpy as np
import matplotlib.pyplot as plt

# 假設 subjects 和 students 已定義
# subjects = ['Math', 'English', 'Science', 'History', 'Art']
# students = np.array([[79, 93, 69, 96, 92],
#                      [66, 76, 64, 85, 87]])

x = np.arange(len(subjects))  # x 軸位置
width = 0.35  # 長條寬度

figure = plt.figure()
axes = figure.add_subplot()

# 畫兩組長條，位置錯開避免重疊
bars1 = axes.bar(x - width/2, students[0], width, label='stu1')
bars2 = axes.bar(x + width/2, students[1], width, label='stu2')

axes.set_xticks(x)
axes.set_xticklabels(subjects)
axes.legend()
axes.set_title("stu1 vs stu2")
axes.set_xlabel("subject")
axes.set_ylabel('scores')

plt.show()
plt.close()
```
