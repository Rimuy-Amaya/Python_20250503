```python
import matplotlib.pyplot as plt

labels = ["Nokia", "Samsung", "Apple", "Lumia"]
values = [20, 30, 45, 10]
colors = ["gray","green","red","blue"]
explode = (0,0,0.1,0)

_,_,autotexts = plt.pie (values,labels=labels,colors=colors,explode=explode,shadow=True,startangle=-90,autopct="%1.1f%%")
plt.setp(autotexts,size=14,color="white")
plt.savefig("test1.jpeg") #儲存圖檔，生成檔案
plt.show()
```
